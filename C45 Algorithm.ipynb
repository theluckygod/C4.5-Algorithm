# Giới thiệu thuật toán
### Thuật toán học C4.5
Thuật toán học C4.5 là thuật toán học tạo ra Cây quyết định được phát triển bởi Ross Quinlan, thường được sử dụng trong lĩnh vực Khai thác dữ liệu. Thuật toán có thể được dùng trên các tập dữ liệu có biến liên tục và biến rời rạc, kết quả phân loại đầu ra là giá trị rời rạc.

Thuật toán học C4.5 là cải tiến dựa trên thuật toán ID3, khác biệt ở phần đưa các biến liên tục trở thành các biến rời rạc bằng cách chọn ra một ngưỡng T sao cho tách biệt thuộc tính về 2 phần:
- Phần 1 có giá trị trên ngưỡng T.
- Phần 2 có giá trị bé hơn hoặc bằng T.

### Các trường hợp cơ bản xảy ra khi xây dựng cây
- Không còn dữ liệu để phân loại, tạo nút lá trả về phân loại của cây cha
- Tất cả các mẫu trong danh sách thuộc cùng một lớp. Khi điều này xảy ra, chỉ đơn giản tạo một nút lá cho cây quyết định chọn lớp đó.
- Không còn thuộc tính nào cung cấp thông tin. Trong trường hợp này, C4.5 tạo một nút lá cho cây quyết định lấy lớp chiếm đa số.
- Trường hợp gặp dữ liệu chưa từng thấy trước đây. Một lần nữa, C4.5 tạo một nút quyết định cho cây , sử dụng giá trị chiếm đa số và tiếp tục phân loại bằng thuộc tính mong đợi.

### Xử lý thuộc tính rời rạc và thuộc tính liên tục
Trong quá trình xây dựng cây, ta sẽ phân thành 2 trường hợp:
- Nếu gặp thuộc tính rời rạc: 
    + $$ Gain = Entropy(Dataset) - AverageEntropy(Dataset, Attribute) $$
- Nếu gặp thuộc tính liên tục: 
    + Sắp xếp dữ liệu với cột thuộc tính này
    + Với mỗi giá trị ***i*** thuộc [1, ***N***], chọn $$ threshold\_temp = (value[i] + value[i - 1])/2 $$, ***value[i]*** là giá trị của dữ liệu ở dòng ***i*** và thuộc tính ***attribute***. Tính độ lợi thông tin với ***threshold_temp*** này. $$ Gain\_temp = Entropy(Dataset) - AverageEntropy(Dataset, Attribute, threshold\_temp) $$
    + Tìm ***threshold_temp*** mà có ***Gain_temp*** lớn nhất, ta có được $$ local\_threshold = maxavg(Gain\_temp) $$
    + Tính độ lợi thông tin với ***local_threshold*** vừa tìm được: $$ Gain = Entropy(Dataset) - AverageEntropy(Dataset, Attribute, local\_threshold) $$

### Cắt tỉa cây sau khi xây dựng
Như đã biết, thuật toán xây dựng cây quyết định thường bị vấn đề overfitting làm cho dự đoán kém hiệu quả khi khai triển trên những dữ liệu chưa từng gặp. Thuật toán C4.5 cải thiện vấn đề này bằng cách cắt tỉa bớt các nút trên cây không cần thiết, và biến các nút này thành nút lá. Quá trình cắt tỉa này có thể được thực hiện trên validation dataset sau quá trình dựng cây. Tiêu chí để đánh giá một node là không cần thiết theo công thức sau:
- $$ ClassificationError(parentNode) > \sum(ClassificationError(subNode)) + getHeight(parentNode) - 1 $$ 


### Những cải tiến từ ID3
- Xử lý cả hai thuộc tính liên tục và rời rạc. Để xử lý các thuộc tính liên tục, thuật toán C4.5 tạo ra một ngưỡng và sau đó chia danh sách thành các thuộc tính có giá trị cao hơn ngưỡng và các thuộc tính nhỏ hơn hoặc bằng nó.
- Xử lý dữ liệu huấn luyện với các giá trị thuộc tính bị thiếu - C4.5 cho phép các giá trị thuộc tính được đánh dấu là '?' cho các giá trị thiếu. Các giá trị thiếu này sẽ không được sử dụng trong tính toán độ lợi thông tin.
- Xử lý các thuộc tính với chi phí khác nhau, các thuộc tính liên tục được xử lý phức tạp hơn các thuộc tính rời rạc, song, hiệu quả phân loại kém hơn rất nhiều so với các thuộc tính này.
- Cắt tỉa cây sau khi tạo: thuật toán C4.5 quay trở lại cây sau khi tạo và cố gắng loại bỏ các nhánh không có ích bằng cách thay thế chúng bằng các nút lá.

# Decision Tree C4.5

```python
class Tree:
    NODE = 0
    LEAF = 1
    
    def __init__(self, predict, type_tree):
        self.predict = predict
        self.type = type_tree
    def isLeaf(self):
        return self.type == Tree.LEAF
    def getPredict(self):
        return self.predict
    def getHeight(self):
        pass
    
class Node(Tree):
    def __init__(self, predict, attribute, type_attribute, threshold=0):
        super().__init__(predict, Tree.NODE)
        self.attribute = attribute
        self.type_attribute = type_attribute
        
        if type_attribute == DecisionTree_C45.CONTINUOUS:
            self.threshold = threshold 
        else:
            self.subVal = []
        
        self.subNode = [] # (if continuous node) subNode[0] for value > threshold, subNode[1] for value <= threshold
        
    def addSubValue(self, value):
        self.subVal.append(value)
    def addSubNode(self, node):
        self.subNode.append(node)
    def travesal(self, data):
        if self.type_attribute == DecisionTree_C45.CONTINUOUS: # continuous attribute
            if type(data[self.attribute]) == str:
                return self, False
            if data[self.attribute] <= self.threshold:
                return self.subNode[1], True
            else:
                return self.subNode[0], True
        else: # discrete attribute      
            for index in range(len(self.subVal)):
                if self.subVal[index] == data[self.attribute]:
                    return self.subNode[index], True
                
        return self, False
    def pruningNode(self):
        self.type = Tree.LEAF
    def splitDataBySubNode(self, data, labels):
        data_ = data[data[:, self.attribute] != '?']
        labels_ = labels[data[:, self.attribute] != '?']
        
        if self.type_attribute == DecisionTree_C45.CONTINUOUS: # continuous attribute
            sub_data = np.empty(shape=(2), dtype='O')
            sub_data[0] = data_[data_[:, self.attribute] > self.threshold]
            sub_data[1] = data_[data_[:, self.attribute] <= self.threshold]
            sub_labels = np.empty(shape=(2), dtype='O')
            sub_labels[0] = labels_[data_[:, self.attribute] > self.threshold]
            sub_labels[1] = labels_[data_[:, self.attribute] <= self.threshold]
            
            return self.subNode, sub_data, sub_labels
        else: # discrete attribute      
            sub_data = np.empty(shape=(len(self.subVal)), dtype='O')
            sub_labels = np.empty(shape=(len(self.subVal)), dtype='O')

            for index in range(len(self.subVal)):
                sub_data[index] = data_[data_[:, self.attribute] == self.subVal[index]]
                sub_labels[index] = labels_[data_[:, self.attribute] == self.subVal[index]]
        return self.subNode, sub_data, sub_labels       
    def getHeight(self):
        if self.type == Tree.LEAF:
            return 1
        
        maxHeight = 1
        for index in range(len(self.subNode)):
            if maxHeight < self.subNode[index].getHeight() + 1:
                maxHeight = self.subNode[index].getHeight() + 1
        return maxHeight
    
class Leaf(Tree):
    def __init__(self, predict):
        super().__init__(predict, Tree.LEAF)
    def getHeight(self):
        return 1
```

```python
def Entropy(labels):
    entropy = 0
    n = len(labels)
    classes, classes_count = np.unique(labels, return_counts=True)
    
    for c in range(len(classes)):
        entropy -= classes_count[c]/n * np.log(classes_count[c]/n)
        
    return entropy
        
def InformationGain(data, labels, attribute):
    data_ = data[data[:, attribute] != '?']
    labels_ = labels[data[:, attribute] != '?']
    
    attribute_value, count = np.unique(data_[:, attribute], return_counts=True)
    AverageEntropy = 0
    n = len(data_)
    for value in range(len(attribute_value)):
        AverageEntropy += count[value] / n * Entropy(labels_[data_[:, attribute] == attribute_value[value]])
        
    return Entropy(labels_) - AverageEntropy

def findThreshold(data, labels, attribute):
    d = data[:, attribute]
    index = sorted(range(len(d)), key=d.__getitem__)
    l = labels[index]
    
    n = len(d)
    local_threshold = 0
    max_IG = -10000
    Entropy_labels = Entropy(l)
    
    for i in range(len(d) - 1):
        temp_threshold = (d[i] + d[i + 1])/2.0
        
        d1 = d[d > temp_threshold]
        d2 = d[d <= temp_threshold]
        AverageEntropy = len(d1) / n * Entropy(l[d > temp_threshold]) + len(d2) / n * Entropy(l[d <= temp_threshold])
        IG = Entropy_labels - AverageEntropy
        if IG > max_IG:
            local_threshold = temp_threshold
            max_IG = IG
    return max_IG, local_threshold

class DecisionTree_C45:
    DISCRETE = 0
    CONTINUOUS = 1
    
    def __init__(self, data, labels, type_of_attributes):
        self.type_of_attributes = type_of_attributes;
        self.root_node = self.createDecisionTree(data, labels, np.array(list(range(data.shape[1]))))
        
    def createDecisionTree(self, data, labels, attributes):        
        # compute class frequency
        labs, lab_count = np.unique(labels, return_counts=True)
        
        # Xét trường hợp cơ bản
        if len(labs) == 1 \
            or len(attributes) == 0 \
            or len(labels) <= 5:  # pruning when the number of instances is smaller than 5
            return Leaf(predict=labs[np.argmax(lab_count)])
        
        # tìm thuộc tính có độ lợi thông tin cao nhất
        att_best = attributes[0]
        IG_max = -10000
        att_best_threshold = 0
        
        for att in attributes:
            if self.isContinuousAttribute(att) == True: # continuous attribute
                IG, threshold = findThreshold(data, labels, att)
            else: # discrete attribute
                IG = InformationGain(data, labels, att)
                
            if IG > IG_max:
                IG_max = IG
                att_best = att
                if self.isContinuousAttribute(att) == True: # continuous attribute
                    att_best_threshold = threshold
                
        # tạo node mới tách từ thuộc tính att_best
        remaining_attributes = attributes[attributes != att_best]
        
        if self.isContinuousAttribute(att_best) == True: # continuous attribute
            new_node = Node(predict=labs[np.argmax(lab_count)], attribute=att_best, type_attribute=DecisionTree_C45.CONTINUOUS, threshold=att_best_threshold)
            
            if len(labels[data[:, att_best] > att_best_threshold]) == 0:
                sub_node0 = Leaf(predict=labs[np.argmax(lab_count)])
            else:
                sub_node0 = self.createDecisionTree(data[data[:, att_best] > att_best_threshold], labels[data[:, att_best] > att_best_threshold], remaining_attributes)
            
            if len(labels[data[:, att_best] <= att_best_threshold]) == 0:
                sub_node1 = Leaf(predict=labs[np.argmax(lab_count)])
            else:
                sub_node1 = self.createDecisionTree(data[data[:, att_best] <= att_best_threshold], labels[data[:, att_best] <= att_best_threshold], remaining_attributes)
            
            new_node.addSubNode(sub_node0)
            new_node.addSubNode(sub_node1)
        else: # discrete attribute
            new_node = Node(predict=labs[np.argmax(lab_count)], attribute=att_best, type_attribute=DecisionTree_C45.DISCRETE)
            
            data_ = data[data[:, att_best] != '?']
            labels_ = labels[data[:, att_best] != '?']
            for value in np.unique(data_[:, att_best]):
                sub_node = self.createDecisionTree(data_[data_[:, att_best] == value], labels_[data_[:, att_best] == value], remaining_attributes)
                new_node.addSubValue(value)            
                new_node.addSubNode(sub_node)
        
        return new_node
        
    def isContinuousAttribute(self, attribute):
        return self.type_of_attributes[attribute] == DecisionTree_C45.CONTINUOUS
    
    def predict(self, data):
        tree = self.root_node
        travesal_successful = True
        
        while travesal_successful != False and tree.isLeaf() != True:
            tree, travesal_successful = tree.travesal(data)
        
        return tree.getPredict()
    
    def evaluate(self, data, labels):
        n = data.shape[0]
        count = 0
        for d, l in zip(data, labels):
            if self.predict(d) == l:
                count += 1
        return count/n
    
    def predictAtNode(self, tree, data):
        travesal_successful = True
        
        while travesal_successful != False and tree.isLeaf() != True:
            tree, travesal_successful = tree.travesal(data)
        
        return tree.getPredict()
            
    def getClassificationErrorAtNode(self, tree, data, labels):
        n = data.shape[0]
        sumClassificationErrorAtSubNode = 0
        for d, l in zip(data, labels):
            if self.predictAtNode(tree, d) == l:
                sumClassificationErrorAtSubNode += 1
        heightTree = tree.getHeight()
        return n - sumClassificationErrorAtSubNode, heightTree
        
    def help_pruning(self, tree, data, labels):
        if tree.isLeaf() == True:
            return
        
        ClassificationErrorAtParentNode = len(labels[labels != tree.getPredict()])
        sumClassificationErrorAtSubNode, heightTree = self.getClassificationErrorAtNode(tree, data, labels)
        
        if ClassificationErrorAtParentNode < sumClassificationErrorAtSubNode + heightTree:
            tree.pruningNode()
        else:
            sub_node, sub_data, sub_labels = tree.splitDataBySubNode(data, labels)
            for n in range(len(sub_node)):
                self.help_pruning(sub_node[n], sub_data[n], sub_labels[n])
    
    def pruning(self, data, labels):
        self.help_pruning(self.root_node, data, labels)
```
